```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.ensemble  import RandomForestRegressor
```

```python
numeric_list=['Max Treating Pressure (psi)','Pre Treatment Shut-in Pressure (psi)','Final Frac Gradient (psi/ft)','Flush Volume (bbls) FRESH' ,'Flush Volume (bbls) PRODUCED']
```

```python
numeric_transformer = Pipeline(steps=[('scaler',StandardScaler())])
```

```python
transformer = ColumnTransformer([
    ('num',numeric_transformer,numeric_list)
])
```

```python
X = df_summary.drop('ISIP (psi)',axis=1)
```

```python
y = df_summary['ISIP (psi)']
```

```python
X_train , X_test , y_train, y_test = train_test_split(X,y,
                                                      test_size =0.30,
                                                      random_state=42)
```

```python
print('Data entrenamiento : ',X_train.shape)
print('Data prueba : ',X_test.shape)
```

```python
%%time
linear_regression = LinearRegression()
model_linear_regression = Pipeline([
    ('transformer',transformer),
    ('linear_regression', linear_regression)
])
```

```python
model_linear_regression.fit(X_train,y_train)
```

```python
muestra_datos =X_test.iloc[:10]
```

```python
model_linear_regression.predict(muestra_datos)
```

```python
np.array(y_test.iloc[:10])
```

```python
print('Accuracy  on training set: {:.2f}'.format(model_linear_regression.score(X_test, y_test)))
```

```python
%%time
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor(random_state=42)
model_tree_reg = Pipeline(steps=[
    ('transformer',transformer),
    ('tree_reg',tree_reg)   
])
model_tree_reg.fit(X_train, y_train)
```

```python
print('Accuracy  on training set: {:.2f}'.format(model_tree_reg.score(X_test, y_test)))
```

```python
model_tree_reg.predict(muestra_datos)
```

```python
np.array(y_test.iloc[:10])
```

```python
%%time
forest_reg = RandomForestRegressor(n_estimators=10,random_state=42)
model_forest_reg = Pipeline(steps=[
    ('transformer',transformer),
    ('forest_reg',forest_reg)  
])
model_forest_reg.fit(X_train,y_train)
```

```python
print('Accuracy  on training set: {:.2f}'.format(model_forest_reg.score(X_test, y_test)))
```
